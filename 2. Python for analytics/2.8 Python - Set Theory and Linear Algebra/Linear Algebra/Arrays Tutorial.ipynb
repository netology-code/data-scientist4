```python
x = np.array([1,2,3])
y = np.array([-7,8,9])
```

```python
dot = np.dot(x,y)
dot
```

```python
x_modulus = np.sqrt((x*x).sum())
y_modulus = np.sqrt((y*y).sum())

cos_angle = dot / x_modulus / y_modulus
angle = np.arccos(cos_angle)
```

```python
angle
```

```python
angle * 360 / 2 / np.pi
```
